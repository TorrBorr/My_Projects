```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")

def tokenize_and_stem(text):
    # first tokenize by sentence, then by word to ensure that punctuation is caught as it's own token
    tokens = [word for sent in nltk.sent_tokenize(str(text)) for word in nltk.word_tokenize(sent)]
    tokens = [s.encode('ascii','ignore') for s in tokens] 
    filtered_tokens = []
    # filter out any tokens not containing letters (e.g., numeric tokens, raw punctuation)
    for token in tokens:
        if re.search('[a-zA-Z]', token):
            filtered_tokens.append(str(token))
    stems = [str(stemmer.stem(t)) for t in filtered_tokens]
    stems  = [s.encode('ascii','ignore') for s in stems ] 
    #print(stems)
    return stems
```

```python
def Game_recommend(game):
    recommendation_list = df_stacked_rec[df_stacked_rec['name'] == game]
    Sorted_recommendation_list = recommendation_list.sort_values(by = 'Total_score', ascending=[False])
    Recommendation_table = Sorted_recommendation_list.drop(['name','Similarity_rank','recommendation','Geek_rating','similarity_score','Total_score'],1)
    
    print("You may also like these games:")
    return(Recommendation_table)
```
